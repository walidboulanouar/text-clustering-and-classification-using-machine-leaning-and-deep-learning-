```python
import pandas as pd
df= pd.read_csv(r"results.csv", index_col=[0])
```

```python
# Step 2: Explore the data
df.info()
```

```python
# Step 3: Data preprocessing
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

documents = df['Body'].values.astype("U")

vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)

k = 6
model = KMeans(n_clusters=k, init='k-means++', max_iter=100, n_init=1)
model.fit(features)

df['cluster'] = model.labels_

df
```

```python
df = df.rename(columns={df.columns[0]: 'ID1'})
df
```

```python
clusters = df.groupby('cluster')    

for cluster in clusters.groups:
    f = open('cluster1'+str(cluster)+ '.csv', 'w', encoding='utf8') # create csv file
    data = clusters.get_group(cluster)['Body'] # get title and Body columns
    f.write(data.to_csv(index=False)) # set index to id
    f.close()
data    

```

```python
print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

```

```python
for i in range(k):
    print("Cluster %d:" % i)
    for j in order_centroids[i, :10]: #print out 10 feature terms of each cluster
        print (' %s' % terms[j])
    print('------------')
```
